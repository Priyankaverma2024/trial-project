```python
auto_price=pd.read_csv('/content/drive/My Drive/Automobile_price_data__Raw_.csv')
auto_price.head()
```

```python
auto_price.dtypes
```

```python
cols=['bore','stroke','horsepower','peak-rpm','price']
auto_price[cols]=auto_price[cols].apply(pd.to_numeric,args=('coerce',))
```

```python
auto_price[cols]
```

```python
auto_price.dtypes
```

```python
auto_price.isnull().sum()
```

```python
auto_price.replace('?',np.nan,inplace=True)
```
